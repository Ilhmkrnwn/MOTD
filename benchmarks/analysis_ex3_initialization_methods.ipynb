```python
metrics = ["feature_importance_lda_{}_test",
           "feature_importance_lda_{}_train_i_0",
           "feature_importance_lda_{}_train_i_9",
           "feature_importance_lda_{}_train_i_199",
           "None_{}_test",
           "None_{}_train_i_0",
           "None_{}_train_i_9",
           "None_{}_train_i_199",
          ]
```

```python
# Percentage 
df_sums = pd.DataFrame(init_methods)
for m in metrics:
    sumcolumns = []
    for im in init_methods:
        sumcolumns.append(m.format(im))
    df_sums[m] = df[selector].sum()[sumcolumns].values / df[selector].sum()[sumcolumns].values[0]  
    
df_sums    
```

```python
# Percentage points diff
df_sums = pd.DataFrame(init_methods)
for m in metrics:
    sumcolumns = []
    for im in init_methods:
        sumcolumns.append(m.format(im))
    df_sums[m] = df[selector].sum()[sumcolumns].values - df[selector].sum()[sumcolumns].values[0]   
    
df_sums    
```

```python
# Percentage diff
df_sums = pd.DataFrame(init_methods)
for m in metrics:
    sumcolumns = []
    for im in init_methods:
        sumcolumns.append(m.format(im))
    df_sums[m] = df[selector].sum()[sumcolumns].values / df[selector].sum()[sumcolumns].values[0] - 1  
    
df_sums    
```

```python
# Clean calculation
df_sums = pd.DataFrame(init_methods)
for m in metrics:
        df_sums[m] = 0
df_sums.set_index(0,inplace=True)

for dataset in df.dataset:
    for metric in df_sums.columns:
        for im in init_methods:
            df_sums.loc[im, metric] += df[(df.dataset == dataset)][metric.format(im)].values[0] / df[(df.dataset == dataset)][metric.format("Random_init")].values[0] - 1
            
df_sums = df_sums / len(df.dataset)
df_sums
```

```python
columns = list(df_sums.columns.values)
new = ["method"]
df_sums["method"] = df_sums.index
df_sums = df_sums[new + columns]
```

```python
str1 = "\\hline\r\n"
str1 += "\\multirow{4}{2cm}{" + "average percentage difference" +"}"
for index, row in df_sums.iterrows():
    for element in row:
        if type(element) == str: 
            str1 += " & " + element.split("_")[0] + " & "
        else:
            str1 += ("%.2f" % round(element, 2)).replace("0.",".") + " & "
    str1 = str1[:-2]        
    str1 +=  "\\\\"
    str1 += "\r\n"
    
print(str1)
```

```python
#B97357 brown
#988ED5 violet

colormap = { 
    'iris_input.pd': "#8EBA42",
    'breast_cancer_input.np' : "#FBC15E" ,
    'cars_input.pd' : "#FBC15E",
    'abalone_input.pd': "#348ABD",
    'steel_input.pd': "#348ABD",
    'contraceptive_input.pd' : "#E24A33" ,
    'students_input.pd' : "#81D0DB",
    'banknote_input.pd' : "#81D0DB",
    'bank_input.pd' : "#B97357",
    'adult_input.pd' :"#E24A33",
    'hrss_input.pd' : "#8EBA42",   
    'generated6_input.np' : "grey",
}
datasets1 = [
            "banknote_input.pd", 
            "cars_input.pd", 
            "generated6_input.np",
            "hrss_input.pd",      
            "steel_input.pd",
            "contraceptive_input.pd", 

]
datasets2 = [           
            "iris_input.pd",
            "bank_input.pd",         
            "breast_cancer_input.np",          
            "adult_input.pd", 
            "abalone_input.pd", 
            "students_input.pd",
]
```

```python
for dataset_group in [datasets1, datasets2]:
    displayed_iterations=200
    fig = plt.gcf()
    fig.set_size_inches(10, 4)
    iterations = np.arange(displayed_iterations)
    for dataset in dataset_group:
        accuracies = []
        for i in range(displayed_iterations):
            accuracies.append(df[df.dataset == dataset]["feature_importance_lda_Random_init_train_i_" + str(i)].values[0])
        plt.plot(iterations,accuracies,linewidth=1,linestyle='--',color=colormap[dataset])
        accuracies = []
        for i in range(displayed_iterations):
            accuracies.append(df[df.dataset == dataset]["feature_importance_lda_BGM_init_train_i_" + str(i)].values[0])
        plt.plot(iterations,accuracies,linewidth=1,linestyle='-',color=colormap[dataset],label=dataset.split("_")[0])
    plt.ylabel("training accuracy", fontsize=12)
    plt.xlabel("iteration", fontsize=12)
    l1 = plt.legend(bbox_to_anchor=(1, 1.02), loc="upper left")

    legend_elements = [Line2D([0], [0], color='grey', linestyle='-', label='initialization'),
                       Line2D([0], [0], color='grey', linestyle='--', label='random')]
    ax = plt.gca()
    ax.legend(handles=legend_elements, bbox_to_anchor=(1, -0.02), loc="lower left")
    ax.add_artist(l1)   
    
    #plt.savefig("plots/test.pdf", bbox_inches="tight")
    plt.show()
                                                    #
```

```python
def translator(word):
    if word == "feature_importance_lda":
        return "2D "
    elif word == "None":
        return "f. "
    else:
        return "error"
```

```python
# Save all
for gm in gate_methods:
    for im in init_methods:
        first_set = True
        for dataset_group in [datasets1, datasets2]:
            if first_set:
                first_set = False
                chunk_name = "datasets1"
            else:
                chunk_name = "datasets2"
                
            displayed_iterations=100
            fig = plt.gcf()
            #fig.set_size_inches(4.65, 3.3)
            fig.set_size_inches(4.65, 2.6)
            iterations = np.arange(displayed_iterations)
            for dataset in dataset_group:
                accuracies = []
                for i in range(displayed_iterations):
                    accuracies.append(df[df.dataset == dataset]["{}_Random_init_train_i_".format(gm) + str(i)].values[0])
                plt.plot(iterations,accuracies,linewidth=1.5,linestyle='--',color=colormap[dataset])
                accuracies = []
                for i in range(displayed_iterations):
                    accuracies.append(df[df.dataset == dataset]["{}_{}_train_i_".format(gm,im) + str(i)].values[0])
                plt.plot(iterations,accuracies,linewidth=1.5,linestyle='-',color=colormap[dataset],label=dataset.split("_")[0])
            plt.ylabel("training accuracy", fontsize=12)
            plt.xlabel("iteration", fontsize=12)
            l1 = plt.legend(bbox_to_anchor=(1, 1.04), loc="upper left")

            legend_elements = [Line2D([0], [0], color='grey', linestyle='-', label=translator(gm) + im.split("_")[0] ),
                               Line2D([0], [0], color='grey', linestyle='--', label=translator(gm) + "random")]
            ax = plt.gca()
            ax.legend(handles=legend_elements, bbox_to_anchor=(1, -0.1), loc="lower left")
            ax.add_artist(l1)   

            #plt.savefig("plots/init_methods_{}_{}_{}.pdf".format(gm,im,chunk_name), bbox_inches="tight")
            plt.show()
```
