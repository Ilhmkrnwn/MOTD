```python
gate_methods = ["feature_importance_lr_max" , "None"]
init_methods = ["Random_init","Kmeans_init","KDTmeans_init","BGM_init"]
datasets = df.dataset.unique()
```

```python
str1 = ""
p1 = "placeholder"
for dataset in datasets:            
        for im in init_methods:
            if dataset != p1:
                str1 += "\\hline\r\n"
                str1 += "\\multirow{4}{2cm}{" + dataset.split("_")[0] +"}"
            p1 = dataset
            str1 += " & "
            str1 += im.split("_")[0] + " & "
            for gm in gate_methods: 
                column = [str(gm + "_" + im + "_test")]
                str1 += ("%.2f" % round(df[(df["dataset"] == dataset)][column].values[0][0], 2)).lstrip("0")
                column = [str(gm + "_" + im + "_test_std")]
                str1 += " $\\pm$" + str(round(df[(df["dataset"] == dataset)][column].values[0][0], 2)).lstrip("0") + " & "
                column = [str(gm + "_" + im + "_train_i_0")]
                str1 += ("%.2f" % round(df[(df["dataset"] == dataset)][column].values[0][0], 2)).lstrip("0") + " & "
                column = [str(gm + "_" + im + "_train_i_9")]
                str1 += ("%.2f" % round(df[(df["dataset"] == dataset)][column].values[0][0], 2)).lstrip("0") + " & "
                column = [str(gm + "_" + im + "_train_i_99")]
                str1 += ("%.2f" % round(df[(df["dataset"] == dataset)][column].values[0][0], 2)).lstrip("0") + " & "
            str1 = str1[:-2]
            str1 +=  "\\\\"
            str1 += "\r\n"
print(str1)
```
