```python
import pandas as pd
import numpy as np
import optuna
```

```python
storage = "sqlite:///optuna_results_initialization.sqlite3"
```

```python
attribute = 'user_attrs_test_score' #'user_attrs_test_score','user_attrs_train_score', 'user_attrs_train_score_i_0','user_attrs_train_score_i_9',
studies = [
    "2021.06.18 10:02:04 init",
    "2021.06.18 10:17:32 init",
    "2021.06.18 11:36:15 init",
    "2021.06.18 11:43:31 init",
    "2021.06.18 11:58:15 init",
    "2021.06.18 16:01:25 init", 	
    "2021.06.18 15:46:34 init",	
    "2021.06.18 15:31:49 init",	
    "2021.06.18 15:16:58 init",
]
```

```python
def top50(df, attribute):
    df = df.sort_values(attribute, ascending=False).head(int(len(df)/2))
    return df
```

```python
loaded_study = optuna.load_study(study_name=studies[0], storage=storage)
df = loaded_study.trials_dataframe()
df = top50(df, attribute)
for key, value in loaded_study.system_attrs.items():
    df[key] = value
df.head(1)
```

```python
for study in studies[1:]:
    loaded_study = optuna.load_study(study_name=study, storage=storage)
    df_new = loaded_study.trials_dataframe()
    df_new = top50(df_new, attribute)
    for key, value in loaded_study.system_attrs.items():
        df_new[key] = value
    df = pd.concat([df, df_new])
    df.reset_index(inplace=True, drop=True)
df.shape
```
