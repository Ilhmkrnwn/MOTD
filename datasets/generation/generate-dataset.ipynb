```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
import pickle
```

```python
mean = [0, 5]
cov = [[9, -4], [-4, 2]]  # diagonal covariance
color = []

x, y = np.random.multivariate_normal(mean, cov, 1000).T
for point in y:
    if point > 5:
        color.append(0)
    else:
        color.append(1)
        
x2, y2 = np.random.multivariate_normal([4, 5], cov, 1000).T
for point in y2:
    color.append(2)
x = np.append(x,x2)
y = np.append(y,y2)

cov = [[3, 0], [0, 0.7]]
points=1000
x2, y2 = np.random.multivariate_normal([-6, 2], cov, points).T
color2 = np.repeat(2,points)
x = np.append(x,x2)
y = np.append(y,y2)
color = np.append(color,color2)

cov = [[0.7, 0], [0, 5]]
points=1000
x2, y2 = np.random.multivariate_normal([-2, -1], cov, points).T
color2 = np.repeat(3,points)
x = np.append(x,x2)
y = np.append(y,y2)
color = np.append(color,color2)

cov = [[1, 0], [0, 1]]
points=500
x2, y2 = np.random.multivariate_normal([-6, -2], cov, points).T
color2 = np.repeat(3,points)
x = np.append(x,x2)
y = np.append(y,y2)
color = np.append(color,color2)

cov = [[1, 0], [0, 1]]
points=500
x2, y2 = np.random.multivariate_normal([15, 7.5], cov, points).T
color2 = np.repeat(4,points)
x = np.append(x,x2)
y = np.append(y,y2)
color = np.append(color,color2)


plt.scatter(x, y, s=1, c=color)
plt.axis('equal')
plt.draw()
```

```python
scaler = MinMaxScaler(feature_range=(0,1))
x2 = scaler.fit_transform(x.reshape(-1, 1))
scaler = MinMaxScaler(feature_range=(0,1))
y2 = scaler.fit_transform(y.reshape(-1, 1))
plt.scatter(x2, y2, s=1, c=color)
plt.draw()
```

```python
generated_input = np.column_stack((x2,y2))
generated_target = color
pickle.dump(generated_input, open("../generated6_input.np", "wb"))
pickle.dump(generated_target, open("../generated6_target.np", "wb"))
```
